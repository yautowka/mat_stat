```python
import random
import matplotlib.pyplot as plt
# для равномерного распределения от -3theta до theta
theta = random.uniform(10, 20)
start = -3*theta
end = theta
print(theta)
```

```python
def score(mn, mx):
  return 4*mx+mn
```

```python
from math import inf
from heapq import merge
epoch = 10000 # количество эпох(сколько раз будем добавлять n_plus)
n_plus = 10000 # количество новых точек на каждой эпохе
mx = -inf
mn = inf
res = []
best = []
# прогоняем эпохи
for e in range(epoch):
  for _ in range(n_plus):
    tmp = random.uniform(start, end)
    mx = max(mx, tmp)
    mn = min(mn, tmp)
  res.append(score(mn, mx))
  best.append(theta)
```

```python
print(len(res))
```

```python
plt.plot(res, label='res')
plt.plot(best, label='best')
plt.ylabel('theta')
plt.xlabel('эпоха')
plt.legend()
plt.show()
```

```python
diff = [abs(best[i]-res[i]) for i in range(len(best))]
plt.plot(diff)
plt.ylabel('sup')
plt.xlabel('эпоха')
plt.show()
```
